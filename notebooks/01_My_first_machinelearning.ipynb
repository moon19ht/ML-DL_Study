```python
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7,
                31.0, 31.0, 31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5,
                34.0, 34.0, 34.5, 35.0, 35.0, 35.0, 35.0, 36.0, 36.0, 37.0,
                38.5, 38.5, 39.5, 41.0, 41.0]  # 도미 길이 데이터

bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0,
                450.0, 500.0, 475.0, 500.0, 500.0, 340.0, 600.0, 600.0,
                700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0,
                925.0, 975.0, 950.0]  # 도미 무게 데이터

import matplotlib.pyplot as plt  # 시각화 라이브러리 불러오기

plt.scatter(bream_length, bream_weight)  # 도미 산점도 그리기
plt.xlabel('Length')  # x축 레이블
plt.ylabel('Weight')  # y축 레이블
plt.show()  # 그래프 출력
```

```python
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2,
                12.4, 13.0, 14.3, 15.0]  # 빙어 길이 데이터

smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4,
                12.2, 19.7, 19.9]  # 빙어 무게 데이터

plt.scatter(bream_length, bream_weight)  # 도미 산점도 그리기
plt.scatter(smelt_length, smelt_weight)  # 빙어 산점도 그리기

plt.xlabel('Length')  # x축 레이블
plt.ylabel('Weight')  # y축 레이블
plt.show()  # 그래프 출력
```

```python
length = bream_length + smelt_length  # 도미와 빙어 길이 합치기
weight = bream_weight + smelt_weight  # 도미와 빙어 무게 합치기

fish_data = [[l, w] for l, w in zip(length, weight)]  # (길이, 무게) 쌍으로 데이터 생성

print(fish_data)  # 데이터 출력
```

```python
fish_target = [1] * 35 + [0] * 14  # 도미: 1, 빙어: 0으로 타깃 생성
print(fish_target)  # 타깃 데이터 출력
```

```python
from sklearn.neighbors import KNeighborsClassifier  # k-최근접 이웃 분류기 불러오기
kn = KNeighborsClassifier()  # 분류기 객체 생성
kn.fit(fish_data, fish_target)  # 데이터로 모델 훈련
kn.score(fish_data, fish_target)  # 훈련 데이터 정확도
kn.predict([[30, 600]])  # 길이 30, 무게 600인 생선 예측
print(kn._fit_X)  # 훈련 데이터 출력
```

```python
print(kn._y)  # 훈련 데이터의 타깃 값 출력
```

```python
kn49 = KNeighborsClassifier(n_neighbors=49)  # 이웃 개수 49로 분류기 생성
kn49.fit(fish_data, fish_target)  # 모델 훈련
kn49.score(fish_data, fish_target)  # 훈련 데이터 정확도
print(35/49)  # 훈련 데이터에서 도미 비율
```
